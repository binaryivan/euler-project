### Problem 3. The prime factors of 13195 are 5, 7, 13 and 29. What is the largest prime factor of the number 600851475143 ?

```python
#функция определяет простое ли число
def prime (a):
    if a == 0:
        return 0
    elif a == 1:
        return 0
    else:
        for i in range (2, a):
            if abs(a) % i == 0:
                return 0
                break
        else:
            return 1
```

```python
#функция раскладывает число на простые множители
def sep(a):
    b = 2
    separators = []
    while b <= a:
        if a % b == 0:
            separators.append(b)
            a = a /b
        else:
            b+=1
    return separators
```

```python
sep(100)
```

```python
# функция раскладывает число на простые множители, не понятно чем отличается от функции sep
def primesep (a):
    separators = sep(a)
    primeseparators = []
    while len(separators) != 0:
        if prime(separators[0])==1:
            primeseparators.append(separators[0])
            del separators[0]
        else:
            separators.append(sep(separators[0]))
            del separators[0]
    return primeseparators
```

```python
primesep(600851475143)
```

```python
sep(600851475143)
```

```python
max(primesep(600851475143))
```

### Problem 4. A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 × 99. Find the largest palindrome made from the product of two 3-digit numbers.

```python
def decade(x):
    decade = 0
    while x >= 10:
        x=x/10
        decade+=1
    return(decade)
```

```python
def rotate (p):
    preverse=[]
    for i in range (len(p), 0, -1):
        preverse.append(p[i-1])
    return preverse
```

```python
def palindrome(x):
    if decomposition(x) == rotate(decomposition(x)):
        return 1
    else:
        return 0
```

```python
# функция раскладывающая число на цифры из которого оно составлено, например 157 = 1, 5, 7
def decomposition (x):
    m =[]
    while decade(x) != 0:
        m.append(x // 10 ** decade(x))
        if decade(x) - decade(x - x // 10 ** decade(x) * 10 ** decade(x)) > 1:
            for i in range(1, decade(x) - decade(x - x // 10 ** decade(x) * 10 ** decade(x))):
                m.append(0)
        x = x - x // 10 ** decade(x) * 10 ** decade(x)
    m.append(x)
    return(m)
```

```python
tmp = []
for a in range (1, 1000):
    for b in range (1, 1000):
        if palindrome(a*b) == 1:
            tmp.append(a*b)
max(tmp)
```

### Problem 9. A Pythagorean triplet is a set of three natural numbers, a < b < c, for which, a^2 + b^2 = c^2. For example, 3^2 + 4^2 = 9 + 16 = 25 = 5^2. There exists exactly one Pythagorean triplet for which a + b + c = 1000. Find the product abc.

```python
def chetnoe(x):
    if x%2==0:
        return(1)
    else:
        return(0)
```

```python
def celoe(x):
    if chetnoe(x)==1:
        return(1)
    elif chetnoe(x-1)==1:
        return(1)
    else:
        return(0)
```

```python
z = []
tmp = []
for a in range(1, 1000):
    for b in range(1, a):
        c = a**2+b**2
        if celoe(c**0.5) == 1:
            tmp.append(a)
            tmp.append(b)
            tmp.append(c**0.5)
            z.append(tmp)
            tmp=[]
z
```
